## 3. Exploratory Data Analysis (EDA)

[**WRITE YOUR EXPLANATION HERE.** Introduce the EDA section. State that you will now use visualizations to explore the data's distributions, find outliers, and identify relationships between variables.]

### Distribution of Numerical Variables

[**WRITE YOUR EXPLANATION HERE.** Explain that you are creating histograms to see the distribution of key numerical columns like Age and Total Amount.]

```python
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.histplot(df['Total Amount'], bins=20, kde=True)
plt.title('Distribution of Total Amount')
plt.show()
```

### Analysis of Categorical Variables

[**WRITE YOUR EXPLANATION HERE.** Explain that you are using bar charts to see the counts of different categories in columns like 'Gender' and 'Product Category'.]

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x='Product Category', data=df)
plt.title('Product Category Distribution')
plt.show()
```

### Outlier Detection

[**WRITE YOUR EXPLANATION HERE.** Explain that you are using boxplots to check for outliers in the numerical data. State your decision on whether to keep or remove them and why.]

```python
sns.boxplot(x=df['Total Amount'])
plt.title('Boxplot of Total Amount')
plt.show()
```

### Relationship Between Variables

[**WRITE YOUR EXPLANATION HERE.** Explain that you are now looking for relationships between variables using a scatter plot and a correlation heatmap.]

```python
sns.scatterplot(x='Age', y='Total Amount', data=df)
plt.title('Age vs. Total Amount')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['number'])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
